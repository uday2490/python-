```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
URL="https://in.puma.com/in/en/outlet"
URL
```

```python
r = requests.get(URL) 
data = r.content
print(data)
```

```python
Soup = BeautifulSoup(data,"html.parser")
Soup.prettify 
print(Soup)
```

```python
names = Soup.find_all(class_="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2")
print(names)
```

```python
product_name = []
for i in range(0,len(names)):
    product_name.append(names[i].get_text())
product_name
```

```python
price=Soup.find_all(class_="whitespace-nowrap text-base line-through opacity-50 override:text-puma-black-300 override:opacity-100")
print(price)
```

```python
ori_price=[]
for i in range(0,len(price)):
    ori_price.append(price[i].get_text())
print(ori_price)
```

```python
discount_price=Soup.find_all(class_="whitespace-nowrap text-base text-puma-red font-bold")
print(discount_price)
```

```python
disc_price=[]
for i in range(0,len(discount_price)):
    disc_price.append(discount_price[i].get_text())
print(disc_price)
```

```python
percentage=Soup.find_all(class_="min-w-content tw-1ghqyvy tw-noc9l0 absolute top-0 left-0 z-[5]")
print(percentage)
```

```python
Dsc_Per=[]
for i in range(0,len(percentage)):
    Dsc_Per.append(percentage[i].get_text())
print(Dsc_Per)
```

```python
df=pd.DataFrame({"product name":product_name,"Original price":ori_price,"Discounted price":disc_price,"Discount percentage":Dsc_Per})
print(df)
```
